```python
def linear(data, value):
	"""Return the index of 'value' in 'data', or -1 if it does not occur"""
	# Go through the data list from index 0 upwards
	i = 0
	# continue until value found or index outside valid range
	while i < len(data) and data[i] != value:
		# increase the index to go to the next data value
		i = i + 1
		if data[i] > value:
			i = len(data)
		
	# test if we have found the value
	if i == len(data):
		# no, we went outside valid range; return -1
		return 'does not occur'
	else:
		# yes, we found the value; return the index
		return i
		

#[1,2,3,4,4,6]

def binary(data,value):
	low = 0
	high = len(data)-1
	
	while (high - low) > 0 and data[low] != value:
		mid = int((high+low) / 2)
		print(mid)
	
		if data[mid] == value:
			low = mid
		elif value < data[mid]:
			high = mid-1
		else:
			low = mid+1
	if data[low] == value:
		return low
	else:
		return -1

def binary_pairs(data,value):
	low = 0
	high = len(data)-1
	
	while (high - low) > 0 and data[low][0] != value:
		mid = int((high+low) / 2)
		print(mid)
	
		if data[mid][0] == value:
			low = mid
		elif value < data[mid][0]:
			high = mid-1
		else:
			low = mid+1
	if data[low][0] == value:
		return low
	else:
		return -1
        
		return -1
```
